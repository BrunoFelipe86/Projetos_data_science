# métodos para plotar dados no Pandas

```python
 df2.head()
```

```python
df2.plot.area()
```

```python
df2.plot.area(alpha = 0.5)
```

```python
df2.plot.bar()
```

```python
df2.plot.bar(stacked = True)
```

```python
df1['A'].plot.hist()
```

```python
df1.plot.line( y = 'A') 
# se queremos usar o index no eixo x é somente necessário submeter a coluna que quremos no eixo Y
```

```python
df1.plot.line( y = 'A', figsize=(12,5), lw=1) #figsize,para alterar o tamanho da figura, e lw, para alterar a espessura da linha
```

```python
df1.plot.scatter(x='A', y='B') #um gráfico scatter padrão
```

```python
df1.plot.scatter(x='A', y='B', c='C', cmap = 'inferno') # o parametro C define que os pontos terão cores diferentes. 
```

```python
df1.plot.scatter(x='A', y='B', s= df1['C']*2, figsize =(8,4))
```

```python
df1.plot.box()
```

```python
df2.plot.box()
```
