```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt 
```

```python
%matplotlib inline 
```

```python
x = np.linspace(0, 5, 11)
```

```python
x
```

```python
y = x*x
```

```python
y
```

```python
### - Qual a forma mais simples de plotar um gráfico ?
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x, y, color='red')
plt.xlabel('Eixo X')
plt.ylabel('Eixo y')
plt.title('Título')
plt.show()
```

```python
plt.plot(x, y, 'r--')
plt.xlabel('Eixo X')
plt.ylabel('Eixo y')
plt.title('Título')
plt.show()
```

```python
### - Criando plots com mais de um gráfico
```

```python
plt.subplots( nrows = 1, ncols = 2)
```

```python
plt.subplot( 1, 2, 1)
plt.plot(x, y, 'r--')
plt.subplot( 1, 2, 2)
plt.plot(x, y, 'g*-')
```

```python
## - A forma mais usal de criar gráficos no matplotlib é através da declaração de instancias. 
```

```python
fig = plt.figure()  # - criamos um objeto figura
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # cria um eixo e adiciona a figura
axes.plot(x, y)  # plota os dados no gráfico
axes.set_xlabel('Eixo X')
axes.set_title('Titulo')

```

```python
fig = plt.figure()  
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.3, 0.3]) #adicionando um segundo eixo

axes1.plot(x, y)  
axes1.set_xlabel('Eixo X')
axes1.set_title('Titulo')

axes2.plot(y,x)
```

## - Método plt.subplots 

```python
fig, ax = plt.subplots()
ax.plot(x, x**3, 'b--')
ax.plot(x, x**4, 'g--')
ax.set_xlabel('Eixo x')
ax.set_ylabel('Eixo y')
ax.set_title('Titulo')
```

```python
fig, ax = plt.subplots(nrows = 1, ncols = 2)

ax[0].plot(x, y, 'b')
ax[0].set_title('Title')

ax[1].plot(y, x, 'r')
ax[1].set_title('Title')
```

### - Especificando caracteristicas da figura

```python
fig = plt.figure(figsize =(8,4), dpi=100) # criando uma figura e especificando o tamanho e resolução
```

```python
fig, axes = plt.subplots( figsize = (8,10))
axes.plot(x,y, 'r-', label = 'Dados 1')
axes.plot(y,x, 'g*', label = 'Dados 2')
axes.set_title('Titulo')
axes.legend()
```
