# - Seaborn

Seaborn é uma biblioteca de visualização de dados baseada em matplotlib. 

## - Plots de distruibuições 

Vamos discutir alguns gráficos que nos permitem visualizar a distribuição de um conjunto de dados

```python
import seaborn as sns
```

Por padrão o seaborn já possue algun Datasets internos que podem ser usados para testar a biblioteca

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.displot(tips['total_bill'])
```

```python
sns.histplot(tips['total_bill'], bins = 20)
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips)
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind = 'reg', color = 'green')  # regressão linear
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind = 'hex', color = 'red') # mapa de calor
```

```python
sns.pairplot(tips)  # pega as variaveis númericas e faz jointplot de cada uma delas
```

```python
sns.pairplot(tips, hue = 'sex')  # com o parâmetro hue é adicionado cores diferentes para categorias diferentes.
                                 # escolhendo 'sex' os dados vão adicionar cores diferentes par homens e mulhres
```

```python
sns.rugplot(tips['total_bill'])
sns.kdeplot(tips['total_bill'])
```

## - Plots categóricos

```python
import numpy as np
```

```python
tips.head()
```

```python
# dados categoricos, dados que não são númericos
```

```python
sns.barplot(x='sex', y = 'total_bill', data=tips, estimator = np.std) # passamos como parametros 
                                                                      # sex no eixo X
                                                                      # total_bill no eixo Y
                                                                      # tips é o nosso dataset
                                                                      # estimator é alguma função matematica, foi usado o método
                                                                      # de desvio padrão do numpy 
```

```python
sns.boxplot(x = 'day', y = 'total_bill', data = tips)
```

```python
sns.boxplot(x = 'total_bill', y = 'day', data = tips, orient = 'h')
```

```python
sns.violinplot(x='day', y = 'tip', data = tips)
```

```python
sns.violinplot(x='day', y = 'tip', data = tips, hue = 'sex', split = True)
```

```python
sns.swarmplot(x = 'day', y ='total_bill', data = tips, hue = 'sex', dodge = True)
```

```python
sns.catplot( x = 'sex', y = 'total_bill', data = tips, kind = 'bar') #usando o método catplot criamos os outros tipos
                                                                        # de gráficos através do parâmetro kind
```

## - Plots matriciais. 

```python
flights = sns.load_dataset('flights')
```

```python
flights.head()
```

```python
## É importante que os nossos dados estejam na forma matricial. 
```

```python
crr = tips.corr() # usando correlação para os valores do dataset tips
```

```python
crr
```

```python
sns.heatmap(crr) # criando um mapa de calor da relação de covariancia do dataset tips
```

```python
sns.heatmap(crr, cmap = 'coolwarm') # parametro cmap muda a paleta de cores
```

```python
sns.heatmap(crr, cmap = 'coolwarm', annot = True) # parametro annot adiciona valores aos quadros
```

```python
flights.info()  #método info fornecerá informações sobre o dataset
```

```python
pf = flights.pivot_table(values = 'passengers', index = 'month', columns = 'year') #método pivot_table é usado para mudar quem
                                                                                    # quem são as colunas, indeces e valores
```

```python
pf
```

```python
sns.heatmap(pf)
```

```python
sns.heatmap(pf, cmap = 'coolwarm', linecolor ='white', linewidths = 1) # podemos adicionar os parametros na visualização do mapa
```

```python
sns.clustermap(pf) # clustermap cria cluster, ou seja, agrupa dados em conjunstos 
```

# - Plots de Regressão 

Vamos usar o seaborn para criar plots de modelos lineares. 

```python
sns.lmplot( x = 'total_bill', y = 'tip', data = tips)
```

```python
sns.lmplot( x = 'total_bill', y = 'tip', data = tips, hue ='sex', markers = ['o','v'], scatter_kws = {'s':100})
```

```python
sns.lmplot( x = 'total_bill', y = 'tip', data = tips, hue ='sex', col = 'sex') # separara o grafico em dois usando o parametro
                                                                               # escolhido em hue
```

## - PairGrids 

```python
iris = sns.load_dataset('iris') # esse é um dataset com dados sobre flores.
import matplotlib.pyplot as plt
```

```python
iris.head(20)
```

```python
g = sns.PairGrid(iris)
g.map(plt.scatter)
```

```python
g = sns.PairGrid(iris)
g.map_diag(plt.hist)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
sns.pairplot(iris, hue = 'species')  
```

```python
f = sns.FacetGrid(tips, col='time', row = 'smoker')
f.map(plt.hist, 'total_bill')
```

## - Estilos e Cores

seaborn conta com uma série de estilos padrões que podem ser usado par modificar os seus gráficos. 

```python
sns.countplot(x='sex', data = tips)
```

```python
sns.set_style('darkgrid')
sns.countplot(x='sex', data = tips)
```

```python
sns.set_style('whitegrid')
sns.countplot(x='sex', data = tips)
```

```python
sns.set_style('ticks')
sns.countplot(x='sex', data = tips)
```

```python
sns.set_style('ticks')
sns.countplot(x='sex', data = tips)
sns.despine()
```

```python
sns.set_style('ticks')
plt.figure(figsize = (10, 5))
sns.countplot(x='sex', data = tips)
```

```python
sns.set_style('darkgrid')
sns.set_context('poster')
sns.countplot(x='sex', data = tips)
```
